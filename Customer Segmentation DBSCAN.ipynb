```python
from scipy.stats import pointbiserialr
from sklearn.preprocessing import LabelEncoder

def point_biserial_matrix(df, categorical_cols, numerical_cols):
    pb_matrix = pd.DataFrame(index=numerical_cols, columns=categorical_cols)
    
    # Lakukan encoding pada setiap kolom kategorikal secara independen
    for cat_col in categorical_cols:
        # Label encode data kategorikal
        le = LabelEncoder()
        encoded_cat = le.fit_transform(df[cat_col].astype(str))
        
        # Hitung korelasi point-biserial antara setiap kolom kategorikal dan numerik
        for num_col in numerical_cols:
            corr, _ = pointbiserialr(encoded_cat, df[num_col])
            pb_matrix.loc[num_col, cat_col] = corr
    
    return pb_matrix.astype(float)

# Hitung korelasi antara variabel kategorikal dan numerikal
pb_matrix = point_biserial_matrix(df, categorical_cols, numerical_cols)

# Visualisasi heatmap Point-Biserial
plt.figure(figsize=(10, 8))
sns.heatmap(pb_matrix, annot=True, cmap='coolwarm')
plt.title('Point-Biserial Correlation Heatmap (Categorical-Numerical)')
plt.show()
```

### DBSCAN CLUSTERING

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from itertools import product

eps =  np.arange(30, 40, 1)
min_sample = np.arange(20,30)
numeric_cols = df[['Age', 'Work_Experience', 'Family_Size']]

# Perform DBSCAN clustering
dbscan_params = list(product(eps, min_sample))
print(dbscan_params)
# # Scale features
# scaler = StandardScaler()
# X_scaled = scaler.fit_transform(df[['Age', 'Work_Experience', 'Family_Size']])

# # Get cluster labels
# # labels = dbscan.labels_
```

```python
from sklearn.metrics import silhouette_score

no_of_clusters = []
sil_score = []

for i in dbscan_params:
    dbs_clustering = DBSCAN(eps=i[0], min_samples=i[1]).fit(numeric_cols)
    # labels = dbs_clustering.labels_
    n_clusters = len(np.unique(dbs_clustering.labels_))
    no_of_clusters.append(n_clusters)
    if n_clusters > 1:
        sil_score.append(silhouette_score(numeric_cols, dbs_clustering.labels_))
    else:
        sil_score.append(-1)
```

```python
tmp = pd.DataFrame.from_records(dbscan_params, columns =['Eps', 'Min_samples'])   
tmp['No_of_clusters'] = no_of_clusters

pivot_1 = pd.pivot_table(tmp, values='No_of_clusters', index='Min_samples', columns='Eps')

fig, ax = plt.subplots(figsize=(12,6))
sns.heatmap(pivot_1, annot=True,annot_kws={"size": 20}, cmap="YlGnBu", ax=ax)
ax.set_title('Number of clusters')
plt.show()
```
