Kolom Profession

Kolom Gender

```python
print('Kolom Gender :', df['Gender'].unique())
print('Modus Kolom Gender :', df['Gender'].mode())
print(df['Gender'].describe())
df['Gender'].value_counts().plot(kind='bar', color='brown')
```

Kolom Ever Married

```python
# Cek Missing Value

print('Kolom Ever Married :', df['Ever_Married'].unique())
print('Modus Kolom Ever Married :', df['Ever_Married'].mode())
print(df['Ever_Married'].describe())
df['Ever_Married'].value_counts().plot(kind='bar', color='brown')
```

```python
# Handling missing value menggunakan modus karena data kategorik
df['Ever_Married'] = df['Ever_Married'].fillna(df['Ever_Married'].mode()[0])
print(df['Ever_Married'].describe())
df['Ever_Married'].value_counts().plot(kind='bar', color='brown')
```

Kolom Age

```python
print(df['Age'].describe())
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
sns.histplot(df['Age'], bins=15, ax=axes[0], kde=True, color='brown')
axes[0].set_title('Histogram of Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')

sns.boxplot(x=df['Age'], ax=axes[1])
axes[1].set_title('Boxplot of Age')
axes[1].set_xlabel('Age')

plt.tight_layout()
plt.show()
```

```python
# Menghitung Q1 (kuartil pertama) dan Q3 (kuartil ketiga)
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)

# Menghitung IQR (Interquartile Range)
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mendapatkan outlier
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

print("Outlier:")
outliers['Age'].unique()
```
